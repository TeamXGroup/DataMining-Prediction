```python
indicateurs = [ 'Consommation intermédiaire','Excédent brut d\'exploitation',  'Impôts sur les produits', 'Production nationale', 'Salaires bruts']
for indicateur in indicateurs:
    plt.figure(figsize=(10, 5))
    for secteur in df['Secteur'].unique():
        data_secteur = df[df['Secteur'] == secteur]
        plt.plot(data_secteur['Année'], data_secteur[indicateur], label=secteur, marker='o')

    plt.title(f' {indicateur} ')
    plt.xlabel('année')
    plt.ylabel(indicateur)
    plt.legend(title='Secteur', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(True)
    plt.show()
```

```python
columns = ['Capitalisation boursière', 'Consommation Finale', 'Consommation intermédiaire','Excédent brut d\'exploitation', 'Exportations', 'Importations', 'Impôts sur les produits','PIB', 'Production nationale', 'Salaires bruts', 'Volume des échanges']
correlation_matrix = df[columns].corr()
plt.figure(figsize=(10,5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation entre les indicateurs')
plt.show()
```
